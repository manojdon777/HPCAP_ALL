```python
from sklearn import datasets
RANDOM_STATE=24
np.random.seed(RANDOM_STATE)
data = datasets.make_blobs(n_samples=50, n_features=2,centers=2,random_state=RANDOM_STATE,cluster_std=0.6)
```

```python
X,Y=data
X.shape, Y.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
clf = SVC(kernel='linear')
clf.fit(X,Y)
y_pred = clf.predict(X)
accuracy_score(Y,y_pred)
```

```python
clf.support_vectors_
```

```python
clf = SVC(kernel='sigmoid')
clf.fit(X,Y)
y_pred = clf.predict(X)
accuracy_score(Y,y_pred)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100, edgecolors='r', alpha=0.6)
```

```python
clf = SVC(kernel='rbf')
clf.fit(X,Y)
y_pred = clf.predict(X)
accuracy_score(Y,y_pred)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100, edgecolors='r', alpha=0.6)
```

```python
for c in [0.001, 1, 1000]:
    for k in ['linear','rbf','poly']:
        for g in [0.001, 1, 'auto']:
            clf = SVC(kernel='rbf', C=c, gamma=g)
            clf.fit(X,Y)
            y_pred = clf.predict(X)
            print(f'C: {c}, kernel: {k}, gamma: {g}, Acc:{accuracy_score(Y, y_pred): .2}')
```

```python
digits = datasets.load_digits()
print(digits.keys())
```

```python
print(digits.DESCR)
```

```python
X=digits.data
y = digits.target
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
TEST_SIZE = 0.25
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE,
                                                    stratify=y,
                                                    random_state=RANDOM_STATE)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
for k in ['linear','rbf','poly']:
    clf = SVC(kernel=k)
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print(f'Kernel: {k}: - Acc:{accuracy_score(y_test, y_pred): .4}', end='')
    print(f' - f1:{f1_score(y_test, y_pred, average="weighted"):0.4f}')
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,6))
i = 0
for k in ['linear','rbf','poly']:
    clf = SVC(kernel = k)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f'Kernel: {k}: - Acc:{accuracy_score(y_test, y_pred): .4}', end='')
    print(f' - f1:{f1_score(y_test, y_pred, average="weighted"):0.4f}') 
    cm = confusion_matrix(y_test, y_pred)
    print('-'*40)
    ax = axes[i]
    i = i + 1
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=digits.target_names)
    disp.plot(ax=ax, cmap=plt.cm.Blues, colorbar=False)
```

```python
X, y = datasets.make_blobs(n_samples=300, centers=4, random_state=RANDOM_STATE, cluster_std=0.60)
fig, ax = plt.subplots()
ax.scatter(X[:, 0],X[:, 1], c=y, s=80, edgecolors='black')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=10)
clf.fit(X,y)
```

```python
data = datasets.load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size= TEST_SIZE, stratify=data.target, random_state=RANDOM_STATE)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf = DecisionTreeClassifier(random_state=RANDOM_STATE)
clf.fit(X_train,y_train)
```

```python
clf.score(X_train, y_train), clf.score(X_test, y_test)
```

```python
tree=DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)
tree.fit(X_train, y_train)
print("Accuracy on train set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on train set: {:.3f}".format(tree.score(X_test, y_test)))
      
```
