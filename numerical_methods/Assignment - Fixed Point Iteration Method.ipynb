```python
import math
def f(x):
    return x**3 - x**2 - 1
def g(x):
    return 1/math.sqrt(1+x)
def fixed_point_iteration(ini_guess, epsilon, iterations):
    
    for i in range(iterations):
        next_guess = g(ini_guess)
        print(f'{next_guess}')
        if abs(f(next_guess)) < epsilon:
            print('break out')
            break

        ini_guess = next_guess
    return next_guess
print(fixed_point_iteration(2,.00001,10))
```
