```python
import matplotlib.pyplot as plt

def f(x):
    return -x*x + 4*x + 5

def f_1st_derivative(x):
    return -2*x + 4

def f_2nd_derivative():
    return -2

x = [i for i in range(-2, 7, 1)]

y = [f(i) for i in x]
plt.plot(y)

y = [f_1st_derivative(i) for i in x]
plt.plot(y)

y = [f_2nd_derivative() for i in x]
plt.plot(y)

plt.show()
```
