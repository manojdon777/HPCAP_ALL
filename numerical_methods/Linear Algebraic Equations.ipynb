```python
import random
import matplotlib.pyplot as plt
slope = 3/5
intercept = 5
x = [random.randint(1,100) for i in range(100)]
y = [(slope * i + intercept) for i in x]

plt.subplot(1,4,1)
plt.plot(x, linestyle = 'dotted', color='Green')
plt.title("X")

plt.subplot(1,4,2)
plt.plot(y, linestyle = 'dotted', color='Green')
plt.title("Y")

plt.subplot(1,4,3)
plt.plot(x,y, linestyle = 'dotted', color='Green')
plt.title("X & Y")

plt.show()
```
